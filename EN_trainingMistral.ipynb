```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import LoraConfig, get_peft_model
import torch
from peft import prepare_model_for_kbit_training



model_id = "mistralai/Mistral-7B-Instruct-v0.2"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16
)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = prepare_model_for_kbit_training(model)

# 🔸 PEFT LoRA Ayarı
lora_config = LoraConfig(
    r=128, #farklı
    lora_alpha=16,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, lora_config)


model.print_trainable_parameters()
```

```python
from datasets import load_dataset

DATA_DIR = "/content/drive/MyDrive/Mixtral_ErrorDataset"
all_data_path = f"{DATA_DIR}/Data_for_Mistral_cleaned.jsonl"
CACHE_DIR = "/content/drive/MyDrive/hf_dataset_cache"
os.makedirs(CACHE_DIR, exist_ok=True)

dataset = load_dataset("json", data_files={"data": all_data_path}, split="data")
dataset = dataset.train_test_split(test_size=0.1, seed=42)

```

```python
tokenizer.pad_token = tokenizer.eos_token
model.config.pad_token_id = tokenizer.eos_token_id

def preprocess(example):
    instruction_text = f"Error log: {example['input']} Analyze the error log above and provide a brief explanation and possible solution."
    response_text = example['output']
    full_conversation_text = f"<s>[INST] {instruction_text} [/INST] {response_text}</s>"
    tokenized = tokenizer(full_conversation_text, truncation=True, max_length=2048, padding="max_length")
    tokenized["labels"] = tokenized["input_ids"].copy()
    return tokenized

tokenized_dataset = dataset.map(preprocess, batched=False)



```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from peft import PeftModel
import torch
import os

# === MODEL YOLLARI ===
BASE_MODEL = "mistralai/Mistral-7B-Instruct-v0.2"
MODELS_PATH = "/content/drive/MyDrive/Mixtral_ErrorDataset/mixtral_lora_model/mixtral_lora_model_1e"
USED_MODEL_FOLDER = ""  # Eğer model direkt bu klasördeyse
ADAPTER_PATH = os.path.join(MODELS_PATH, USED_MODEL_FOLDER)

# === DONANIM SEÇİMİ ===
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Model {device} üzerinde yüklenecek.")

try:
    # === QUANTIZATION AYARLARI ===
    bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16)

    # === MODEL VE TOKENIZER YÜKLE ===
    model = AutoModelForCausalLM.from_pretrained(
        BASE_MODEL,
        quantization_config=bnb_config,
        device_map="auto"
    )

    tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL)  # Mistral tokenizer
    model = PeftModel.from_pretrained(model, ADAPTER_PATH)

    print("Baz model, tokenizer ve adaptör katmanları başarıyla yüklendi!")

except Exception as e:
    print(f"Model veya tokenizer yüklenirken bir hata oluştu: {e}")
    print(f"Lütfen ADAPTER_PATH = {ADAPTER_PATH} klasörünün doğru olduğundan emin olun.")

# === Tokenizer için padding token ayarı ===
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
    print(f"Tokenizer'ın pad_token'ı {tokenizer.pad_token} olarak ayarlandı.")

# === INFERENCE HAZIRLIĞI ===
MAX_TOTAL_TOKENS = 4096
MAX_TOKENS_FOR_LOG = 2048  # Örnek olarak ayrılacak hata log token sayısı

PROMPT_TEMPLATE = (
    "Please analyze the error log above.\n"
    "Provide a concise but professional explanation of the error.\n"
    "Then suggest a practical, developer-friendly solution.\n"
    "💡 Explanation:\n"
    "🔧 Solution:"
)

# === ERROR LOG ÖN İŞLEME ===
def preprocess_error_log(log_text, max_tokens_for_log):
    lines = log_text.split('\n')
    processed_lines = []

    found_main_error_message = False
    my_app_namespaces = ["myapplication", "mywebapp", "mycompany"]
    stack_lines_count = 0

    for line in lines:
        line_lower = line.lower()

        if ("system.data.sqlclient.sqlexception" in line_lower or
            "an unhandled exception occurred" in line_lower or
            "system.outofmemoryexception" in line_lower or
            "system.nullreferenceexception" in line_lower or
            "system.io.ioexception" in line_lower or
            "system.invalidoperationexception" in line_lower):
            processed_lines.append(line)
            found_main_error_message = True

        elif any(ns in line_lower for ns in [f"at {ns}" for ns in my_app_namespaces]):
            if stack_lines_count < 5:
                processed_lines.append(line)
                stack_lines_count += 1

        elif "--- end of inner exception stack trace ---" in line_lower:
            processed_lines.append(line)
            break

        elif not found_main_error_message and len(processed_lines) < 3:
            processed_lines.append(line)

    short_log_text = "\n".join(processed_lines)
    tokenized_short_log = tokenizer.encode(short_log_text, add_special_tokens=False)

    if len(tokenized_short_log) > max_tokens_for_log:
        truncated_log = tokenizer.decode(tokenized_short_log[:max_tokens_for_log], skip_special_tokens=True)
        return truncated_log
    return short_log_text
```

```python
def analyze_error_log(error_log: str):
    # 1. Hata logunu ön işle
    processed_log = preprocess_error_log(error_log, max_tokens_for_log=2048)

    # 2. Prompta yerleştir
    prompt = f"<s>[INST] Error log: {processed_log}\n{PROMPT_TEMPLATE} [/INST]"

    # 3. Tokenize et
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    # 4. Modelden yanıt al
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=512,
            temperature=0.7,
            top_p=0.9,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id
        )

    # 5. Cevabı çöz ve sadece yanıt kısmını al
    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
    response = decoded.split("[/INST]")[-1].strip()
    return response
```

```python
while True:
    print("\n--------------------------------------------------")
    user_log = input("Lütfen hata logunu yapıştırın (çıkmak için 'exit' yazın):\n")
    if user_log.lower() == 'exit':
        break

    print("\nModel Analiz Ediyor...")
    analysis_result = analyze_error_log(user_log)
    print(f"\nModelin Analizi:\n{analysis_result}")
    print("--------------------------------------------------\n")
```
